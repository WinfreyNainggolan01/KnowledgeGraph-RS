```python
import pandas as pd

# Read CSV files
wisata_item = pd.read_csv("wisata-item.csv")
wisata_tobawisa = pd.read_csv("wisata_TOBAWISA.csv", delimiter='|')

# Clean rating column (remove comma and convert to float)
wisata_tobawisa['rating'] = wisata_tobawisa['rating'].str.replace(',', '.').astype(float)

# Merge dataframes on item_id
merged = pd.merge(
    wisata_item, 
    wisata_tobawisa[['item_id', 'rating']], 
    on='item_id', 
    how='left'
)

# Save to new CSV
merged.to_csv("wisata-item-updated.csv", index=False)

print("✅ File berhasil digabungkan!")
print(f"\nPreview data:")
print(merged.head())
```

```python
import pandas as pd

# Read the CSV files
restaurant_df = pd.read_csv("restaurant_TOBAREST.csv", delimiter='|')
hotel_df = pd.read_csv("hotel_TOBAHOTL.csv", delimiter='|')

# Read your KG data (assuming it's in a CSV file with item_id and count_review)
# Replace 'kg_data.csv' with your actual KG data file
kg_data = pd.read_csv("facility-item.csv")  # or create DataFrame from your data

# Clean rating column for restaurant (remove comma and convert to float)
restaurant_df['rating'] = restaurant_df['rating'].str.replace(',', '.').astype(float)

# Clean rating column for hotel (remove comma and convert to float if needed)
hotel_df['rating'] = hotel_df['rating'].str.replace(',', '.').astype(float)

# Combine both dataframes
combined_df = pd.concat([
    restaurant_df[['item_id', 'rating']],
    hotel_df[['item_id', 'rating']]
], ignore_index=True)

# Merge with KG data
result = pd.merge(
    kg_data,
    combined_df,
    on='item_id',
    how='left'
)

# Save result
result.to_csv("facility-item_with_ratings.csv", index=False)

print("✅ Rating berhasil dimapping!")
print(f"\nPreview data:")
print(result.head(10))
print(f"\nTotal rows: {len(result)}")
print(f"\nMissing ratings: {result['rating'].isna().sum()}")
```
